```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=36, 
    weight_grams=100,
    waste_factor=0.01
)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = 0.35
title='Whole Wheat Multigrain Bread'

formula = create_formula({
    "ww-flour": 50,
    "bread-flour": 50,
    "flax seeds (soaker)": 5,
    "sunflower seeds (soaker)": 5,
    "cracked wheat (soaker)": 5,
    "oats (soaker)": 5,
    "water": 53,
    "soaker water (boiling)": 25,
    "salt": 2.4,
    "yeast": .43,
    "honey": 5
})

pate_fermentee = create_formula({
    "bread-flour": 100,
    "water": 65,
    "salt": 2,
    "yeast": 0.066
})

steps='''
- prepare pate fermentee at least 12-16 hours ahead
- prepare soaker at least 4 hours ahead
- add pate fermentee after other ingredients are mixed
- bulk ferment: 2 hours, fold once
- divide and shape
- proof: 60-90 min
- bake at 450F with steam
'''

pate_fermentee, final_dough = calc.calculate(formula, pate_fermentee_df=pate_fermentee, 
                                             pre_fermented_flour_ratio=pre_fermented_flour_ratio)
format_and_display(formula=final_dough, pate_fermentee=pate_fermentee, calc=calc, title=title, steps=steps)
```
