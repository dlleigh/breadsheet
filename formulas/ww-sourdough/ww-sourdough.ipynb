## Whole Wheat Sourdough



```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=8, 
    weight_pounds=1,
    weight_ounces=12,
    waste_factor=0.02
)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = 0.20
title='Whole Wheat Sourdough'

formula = create_formula({
    "whole-wheat-flour": 100,
    "water": 85,
    "malt": 0.8,
    "salt": 2
})

desem = create_formula({
    "whole-wheat-flour": 100,
    "water": 60,
    "seed": 10
})

steps='''
- Autolyse 15 min
- Bulk fermentation: 3 hr (fold every 15 min for first 90 min)
- shape
- Proof: 90 min'''

desem, final_dough = calc.calculate(formula, desem_df=desem, pre_fermented_flour_ratio=pre_fermented_flour_ratio)
format_and_display(formula=final_dough, desem=desem, calc=calc, title=title, steps=steps, reserved_seed_grams=75)
```
