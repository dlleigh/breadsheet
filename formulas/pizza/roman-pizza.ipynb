## Roman Pizza



```python
import sys
sys.path.append('../..')
from breadsheet_utils import PrefermentCalculator, create_formula, format_and_display
```

```python
calc = PrefermentCalculator(
    num_loaves=1, 
    weight_grams=760,
    waste_factor=0.02
)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio=.17
title='Roman Pizza'

formula = create_formula({
    "bread-flour": 100 * pre_fermented_flour_ratio,
    "high-gluten-flour": 67,
    "whole-wheat-flour": 16,
    "water": 83.5,
    "malt": 0.42,
    "salt": 1.83
})

levain = create_formula({
    "bread-flour": 100,
    "water": 100,
    "levain": 20
})


steps='''
- autolyse 20 min
- mix
- bulk ferment 1 hour, fold 2-3 times
- cold retard 42-78 hours
- shape/pan 2-3 hours before baking
- bake 500 degrees
'''

levain, final_dough = calc.calculate(formula, levain_df=levain, pre_fermented_flour_ratio=pre_fermented_flour_ratio)
format_and_display(formula=final_dough, levain=levain, calc=calc, title=title, steps=steps, reserved_seed_grams=150)
```
