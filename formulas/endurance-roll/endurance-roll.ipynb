## Endurance Roll


```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=24, 
    weight_grams=50,
    waste_factor=0.02
    )
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = .55
hydration = 62

dough = create_formula({
    "high-gluten flour": 70,
    "WW flour": 30, 
    "water": hydration,
    "salt": 2.5,
    "instant yeast": .81,
    "malt": 0.94,
    "molasses": 8,
    "olive oil": 4,
    "almonds": 16,
    "dates": 30,
    "chocolate": 9
})

sponge = create_formula({
    "high-gluten flour": 100,
    "water": hydration / pre_fermented_flour_ratio, 
    "instant yeast": 0.56
})

steps = '''
- mix sponge, rest 2 hr
- mix dough, adding nuts & dried fruit at end of mix
- divide & shape (with chocolate)
- retard: 8-36 hours
- boil: 30s per side
- bake: 450 degrees, 20 min
'''

title='Endurance Roll'
```

```python
sponge_weights, final_dough = calc.calculate(dough, sponge_df=sponge, pre_fermented_flour_ratio=pre_fermented_flour_ratio)

format_and_display(formula=final_dough, sponge=sponge_weights,
                   calc=calc, title=title, steps=steps)
```
