## Endurance Roll


```python
import sys
sys.path.append('../..')
from breadsheet_utils import PrefermentCalculator, create_formula, format_and_display
```

```python
calc = PrefermentCalculator(
    num_loaves=32, 
    weight_grams=100)
calc.print_batch_info()
```

```python
dough = create_formula({
    "high-gluten-flour": 15,
    "WW Flour": 30, 
    "yeast": 0.5,
    "salt": 2.5,
    "malt": 0.94,
    "molasses": 8,
    "olive oil": 4,
    "almonds": 18,
    "dates/almonds": 16,
    "chocolate": 11
})

sponge = create_formula({
    "high-gluten-flour": 55,
    "water": 62, 
    "yeast": 0.31
})

steps = '''
- mix sponge, rest 2 hr
- mix dough
- divide & preshape: 20 min
- shape
- proof: 20 min (until they float)
- retard: 8-36 hours
- boil: 30s per side
- bake: 450 degrees, 20 min
'''

title='Endurance Roll'
```

```python
sponge_weights, final_dough = calc.calculate_sponge_recipe(sponge_df=sponge,
                                                           dough_df=dough)

format_and_display(formula=final_dough, sponge=sponge_weights,
                   calc=calc, title=title, steps=steps)
```
