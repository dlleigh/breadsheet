## Endurance Roll


```python
import sys
sys.path.append('../..')
from breadsheet_utils import PrefermentCalculator, create_formula, format_and_display
```

```python
calc = PrefermentCalculator(
    num_loaves=12, 
    weight_grams=100)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = .55

dough = create_formula({
    "high-gluten flour": 70,
    "WW Flour": 30, 
    "water": 62,
    "salt": 2.5,
    "yeast (IDY)": .81,
    "malt": 0.94,
    "molasses": 8,
    "olive oil": 4,
    "almonds": 16,
    "dates/almonds": 14,
    "chocolate": 9
})

sponge = create_formula({
    "high-gluten flour": 100,
    "water": 62 / pre_fermented_flour_ratio, 
    "yeast (IDY)": 0.56
})

steps = '''
- mix sponge, rest 2 hr
- mix dough
- divide & preshape: 20 min
- shape
- proof: 20 min (until they float)
- retard: 8-36 hours
- boil: 30s per side
- bake: 450 degrees, 20 min
'''

title='Endurance Roll'
```

```python
sponge_weights, final_dough = calc.calculate_sponge_recipe(dough, sponge, pre_fermented_flour_ratio)

format_and_display(formula=final_dough, sponge=sponge_weights,
                   calc=calc, title=title, steps=steps)
```
