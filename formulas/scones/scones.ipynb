## Scones



```python
import sys
sys.path.append('../..')
from breadsheet_utils import PrefermentCalculator, create_formula, format_and_display
```

```python
calc = PrefermentCalculator(
    num_loaves=24, 
    weight_grams=75
)
calc.print_batch_info()
```

```python
title='Scones'

formula = create_formula({
    "AP-flour": 100,
    "baking-powder": 5,
    "baking soda": 1.25,
    "salt": 1.21,
    "brown-sugar": 29.2,
    "butter": 29.6,
    "vanilla": 1.67,
    "buttermilk": 82.5
})

steps='''
- pre-heat oven to 500 degrees.
- Mix flour, baking powder, baking soda, salt, and sugar. 
- Cut in butter. Add berries, nuts, flavorings as desired. 
- Add buttermilk and knead to form stiff dough.
- Shape to about 1 inch thick and cut into 8 sections.
- Bake for 8 minutes.  
'''

final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps)
```
