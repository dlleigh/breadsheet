```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=8, 
    weight_grams=70
)
calc.print_batch_info()
```

```python
title='Scones'

formula = create_formula({
    "AP-flour": 100,
    "baking-powder": 4.8,
    "baking soda": 0.8,
    "salt": 1.2,
    "butter": 27,
    "buttermilk": 80
})

steps='''
- pre-heat oven to 500 degrees.
- Mix flour, baking powder, baking soda, and salt. 
- Cut in butter.
- Add buttermilk and knead to form stiff dough.
- Shape to about 1 inch thick and cut into 8 sections.
- Bake for 8 minutes.  
'''

final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps,show_volume=True)
```
