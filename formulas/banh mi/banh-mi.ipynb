## Banh Mi Baguettes

```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
formula = create_formula({
    "bread-flour": 70,
    "high-gluten-flour": 30,
    "water": 70,
    "sugar": 1.5,
    "salt": 2,
    "yeast": 1.4,
    "oil": 2,
    "malt": 1
})

steps= '''
- bulk ferment 1 hr at 75F
- retard overnight
- shape as mini-baguettes
- proof 1 hr
- bake at 450, 20-25 min with steam
'''
title="Banh mi baguettes"
```

```python
calc = RecipeCalculator(
    num_loaves=40, 
    weight_grams=150
)
calc.print_batch_info()
```

```python
final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps)
```
