```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=36, 
    weight_grams=50,
    waste_factor=0.01)
calc.print_batch_info()
```

```python
formula = create_formula({
    "ap-flour": 52, 
    "ww-flour": 48,
    "water": 52,
    "salt": 2,
    "ghee (melted)": 7.7
})
title = "Paratha"
steps = '''
- rest dough 30 min
- divide and roll into thin circles
- (optional) sprinkle with ajwain seeds
- brush with ghee, fold, brush, fold and roll again
- cook on hot griddle with ghee until golden brown
- wrap in foil if not serving immediately
'''
```

```python
final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps)
```
