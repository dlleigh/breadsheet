## Brioche

```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
formula = create_formula({
    "bread-flour": 77,
    "high-gluten-flour": 23,
    "water": 9.6,
    "egg": 49,
    "butter": 50,
    "sugar": 12,
    "salt": 2.5,
    "yeast": 1.65,
})
steps= '''
- soften butter and add to mix after gluten development
- bulk ferment 1 hr, fold, refrigerate
- shape
- proof
- bake at 380
'''
title="Brioche"
```

```python
calc = RecipeCalculator(
    num_loaves=30, 
    weight_grams=80,
    waste_factor=0.02,
)
calc.print_batch_info()
```

```python
final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps)
```
