## Sourdough Semolina
From _Bread_ by Jeffrey Hamelman

```python
import sys
sys.path.append('../..')
from breadsheet_utils import PrefermentCalculator, create_formula, format_and_display
```

```python
calc = PrefermentCalculator(
    num_loaves=36, 
    weight_grams=100,
    waste_factor=0.02
)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = 0.15
title='Semolina Bread'

formula = create_formula({
    "bread-flour": 40,
    "durham-flour": 60,
    "water": 67,
    "malt": 1,
    #"sesame seeds": 5,
    "salt": 2
})

levain = create_formula({
    "bread-flour": 100,
    "water": 100,
    "seed": 20
})
steps='''
- Bulk fermentation: 120 min (fold once)
- Proof: 2 hrs
- Bake: 460F with steam
'''

levain, final_dough = calc.calculate(formula, levain_df=levain, pre_fermented_flour_ratio=pre_fermented_flour_ratio)
format_and_display(formula=final_dough, levain=levain, calc=calc, title=title, steps=steps, reserved_seed_grams=150)
```
