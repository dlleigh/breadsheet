## Sourdough



```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=8, 
    weight_pounds=1,
    weight_ounces=12
)
calc.print_batch_info()
```

```python
pre_fermented_flour_ratio = 0.1667
title='Sourdough'

formula = create_formula({
    "bread-flour": 90,
    "whole-wheat-flour": 10,
    "water": 71.8,
    "malt": 0.83,
    "salt": 2.0
})

levain = create_formula({
    "bread-flour": 100,
    "water": 100,
    "seed": 20
})

steps='''
- Autolyse 25 min
- Bulk fermentation: 150 min (fold every 50 min)
- Proof: 2 hrs'''

levain, final_dough = calc.calculate(formula, levain_df=levain, pre_fermented_flour_ratio=pre_fermented_flour_ratio)
format_and_display(formula=final_dough, levain=levain, calc=calc, title=title, steps=steps, reserved_seed_grams=150)
```
