## Mandel Bread



```python
import sys
sys.path.append('../..')
from breadsheet_utils import RecipeCalculator, create_formula, format_and_display
```

```python
calc = RecipeCalculator(
    num_loaves=2, 
    weight_grams=500
)
calc.print_batch_info()
```

```python
title='Mandel Bread'

formula = create_formula({
    "oil": 38,
    "sugar": 23,
    "almond extract": 0.9,
    "egg": 34,
    "ap-flour": 100,
    "baking-powder": .9,
    "salt": .7,
    "almonds": 32
})

steps='''
- Beat together oil, sugar and extract.  
- Beat eggs and add to mixture.  
- Mix together flour, baking powder and salt and add to mixture.  
- Add nuts.  
- Shape into loaves.  
- Bake in 350 degree oven for 30 minutes or until done.
'''

final_dough = calc.calculate_straight_dough(formula)
format_and_display(final_dough, calc, title=title, steps=steps)
```
